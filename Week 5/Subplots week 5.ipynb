```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories = ['Cat A', 'Cat B', 'Cat C', 'Cat D', 'Cat E']
values = [23, 56, 41, 62, 19]

# Create the figure and subplots
fig, axes = plt.subplots(2, 2)

# Plot a bar chart on the first subplot (top-left)
axes[0, 0].bar(categories, values)
axes[0, 0].set_title('Bar Chart')
axes[0, 0].set_xlabel('Categories')
axes[0, 0].set_ylabel('Values')


# Display the figure
plt.show()
```

```python
# Data for the histogram
data = [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10]

# Create the figure and subplots
fig, axes = plt.subplots(2, 2)

# Plot a histogram on the second subplot (top-right)
axes[0, 1].hist(data, bins=5, edgecolor='black')
axes[0, 1].set_title('Histogram')
axes[0, 1].set_xlabel('Data')
axes[0, 1].set_ylabel('Frequency')



# Display the figure
plt.show()
```

```python
# Data for the pie chart
labels = ['Apple', 'Banana', 'Orange', 'Mango']
sizes = [30, 25, 15, 30]

# Create the figure and subplots
fig, axes = plt.subplots(2, 2)

# Plot a pie chart on the third subplot (bottom-left)
axes[1, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Pie Chart')

# Display the figure
plt.show()
```

```python
# Complete excercise in one cell
import matplotlib.pyplot as plt

# Data for the bar chart
categories = ['Cat A', 'Cat B', 'Cat C', 'Cat D', 'Cat E']
values = [23, 56, 41, 62, 19]

# Data for the histogram
data = [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10]

# Data for the pie chart
labels = ['Apple', 'Banana', 'Orange', 'Mango']
sizes = [30, 25, 15, 30]

# Data for the scatter plot
x_values = [1, 2, 3, 4, 5]
y_values = [2, 5, 3, 6, 4]

# Create the figure and subplots
fig, axes = plt.subplots(2, 2)

# Plot a bar chart on the first subplot (top-left)
axes[0, 0].bar(categories, values)
axes[0, 0].set_title('Bar Chart')
axes[0, 0].set_xlabel('Categories')
axes[0, 0].set_ylabel('Values')

# Plot a histogram on the second subplot (top-right)
axes[0, 1].hist(data, bins=5, edgecolor='black')
axes[0, 1].set_title('Histogram')
axes[0, 1].set_xlabel('Data')
axes[0, 1].set_ylabel('Frequency')

# Plot a pie chart on the third subplot (bottom-left)
axes[1, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Pie Chart')

# Plot a scatter plot on the fourth subplot (bottom-right)
axes[1, 1].scatter(x_values, y_values)
axes[1, 1].set_title('Scatter Plot')
axes[1, 1].set_xlabel('X Values')
axes[1, 1].set_ylabel('Y Values')

# Adjust the spacing between subplots
fig.tight_layout()

# Display the figure
plt.show()
```

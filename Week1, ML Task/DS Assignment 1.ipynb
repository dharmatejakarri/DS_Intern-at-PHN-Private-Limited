```python
data_merged = pd.merge(car_data, car_loan_data, on=['Model Name'])
```

```python
data_merged
```

```python
data_merged.fillna(value=0, inplace=True)
```

```python
GST = 0.20

for index, row in data_merged.iterrows():
    gst_amount = row["On road pricing"] * GST
    data_merged.loc[index, "On road pricing"] += gst_amount
```

```python
import pandas as pd
df_car_details = pd.read_csv("car_details.csv")
df_car_pricing = pd.read_excel("car_pricing.xlsx")
```

```python
df_car_details
```

```python
df_car_pricing
```

```python
df_merged = pd.merge(df_car_details, df_car_pricing, on=['Company name'])
```

```python
df_merged
```

```python
# Filling NA values in the merged data frame:
df_merged.fillna(value=0, inplace=True)
```

```python
df_merged
```

```python
# Assume that the GST rate is 18% for the car
GST_RATE = 1.18  

for index, row in df_merged.iterrows():
    on_road_pricing = row['On road pricing']
    on_road_pricing = on_road_pricing * GST_RATE  
    df_merged.at[index, 'On road pricing'] = on_road_pricing   
```
